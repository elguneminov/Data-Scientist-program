### 1. XML data --> https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/student.xml

Save data from this link into your local machine then try to read it

```python
st = et.parse("student.xml")
root = st.getroot()
df_cols = ["id", "name", "salary"]

out_df = pd.DataFrame(columns = df_cols)

for node in root: 

    s_name = node.attrib.get("name")

    s_id = node.find("id").text if node is not None else None

    s_salary = node.find("salary").text if node is not None else None

    

    out_df = out_df.append(pd.Series([s_id, s_name, s_salary],index = df_cols),ignore_index = True)
print(out_df,"\n")
print(out_df.describe(),"\n")
print(out_df.info(),"\n")
```
