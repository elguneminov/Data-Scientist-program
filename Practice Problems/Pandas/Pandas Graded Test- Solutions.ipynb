###  6. Find average price of house respective to bedroom.
example : avg price of house of 1 bedroom then 2 ...so on

```python
bedroom = house_rental['Bedroom'].max()
for i in range(1,bedroom+1):
    h = house_rental[(house_rental['Bedroom']==i)]
    avg_price = h['Price'].mean()
    print(f'Average price of house with {i} Bedroom : {avg_price}')
```

# Data : 2 'Health Care Data'

```python
# link -- > https://github.com/edyoda/data-science-complete-tutorial/blob/master/Data/HealthCareData-1.xlsx
# click on view raw it will download file automatically in your system then you this file as syntax below
hc_df = pd.read_excel('HealthCareData.xlsx')
hc_df.head()
```

### 1. Display all the information respective to gender.

```python
gender = hc_df.groupby(['Gender',"Age"])
gender.describe()
```

### 2. Find all the male who is consuming alcohal more than 5 years and has diabetes result positive.

```python
hc_df[((hc_df['Gender']=='male')&(hc_df['Duration of alcohol consumption(years)']>5)&(hc_df['Diabetes Result']=='YES'))].head()
```

### 3. Find where more people consume alcohol in rural or urban?

```python
rural = hc_df.apply(lambda x: True if x['Place(location where the patient lives)']=='rural' else False, axis=1)
count_rural = len(rural[rural==True].index)
urban = hc_df.apply(lambda x: True if x['Place(location where the patient lives)']=='urban' else False, axis=1)
count_urban = len(urban[urban==True].index)
if count_urban>count_rural:
    print("Urban's people consume more alcohal than rural's people")
else:
    print("Rural's people consume more alcohal than urban's people")
```

### 4. Find average age of male and female who consume alcohol.

```python
female = hc_df[(hc_df['Gender']=='female')]
print(f'Female average age: {female.Age.mean()}')
male = hc_df[(hc_df['Gender']=='male')]
print(f'Male average age: {male.Age.mean()}')
```

### 5. Display distribution for different type of alcohol.

```python
hc_df.groupby(['Type of alcohol consumed']).describe()
```

### 6. Display average age of people who has diabetes positive.

```python
diabetes  = hc_df[(hc_df['Diabetes Result']=='YES')]
print(f'Average age: {diabetes.Age.mean()}')
```

### 7. Find is there any missing value of age in given data.

```python
hc_df[hc_df.Age.isnull()]
```

### 8. If gender is male and age is less than 35 then change diabetes result to NO .

```python
hc_df.loc[((hc_df['Gender']=='female') & (hc_df['Age']<35)),"Diabetes Result"]='NO'
hc_df[hc_df['Age']<35]
```
