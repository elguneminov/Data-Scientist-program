#### COVID 19 Data Analysis

* Loading data & basic descriptive analysis

```python
covid19_confirmed = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/time_series_covid19_confirmed_global.csv')
```

```python
covid19_confirmed.head()
```

```python
covid19_confirmed.sample(10)
```

### Tell something about the data
* Each country can have many Province/State
* Daily confirmed cases will be along the columns

```python
covid19_confirmed.columns
```

```python
covid19_confirmed.info()
```

### We want data combined country wise

```python
covid19_confirmed.drop(columns=['Lat','Long','Province/State'], inplace=True)
```

```python
covid19_confirmed.sample(10)
```

```python
covid19_confirmed_country = covid19_confirmed.groupby(['Country/Region']).sum(axis=1)
#Or
#covid19_confirmed_country = covid19_confirmed.groupby(['Country/Region']).sum()
```

```python
help(covid19_confirmed.groupby(['Country/Region']).sum)
```

```python
covid19_confirmed_country.sample(3)
```

```python
covid19_confirmed_country.loc[['US','China']]
```

```python
covid19_confirmed_countryT = covid19_confirmed_country.T
```

```python
covid19_confirmed_countryT.info()
```

```python
covid19_confirmed_countryT.tail(10)
```

```python
current_status = covid19_confirmed_countryT.loc['3/29/20']
```

```python
top10_infected = current_status.sort_values(ascending=False)[:10]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

* When to use bar graph - Comparing numerical data of different things

```python
top10_infected.plot.bar()
```

* How infaction has spread in US

```python
us_confirmed_data = covid19_confirmed_country.loc['US']
```

```python
us_confirmed_data.plot()
```

* Comparing US, Italy & China

```python
covid19_confirmed_country.loc[['US','China','Italy']].T.plot()
```
