### Data Transformation

* Changing the column 
* Sales & Marketing Team will be merged
* Rename dept column for sales & marketing people to S&M 

```python
def f(d):
    if (d == 'sales') | (d == 'marketing'):
        return 'S&M'
    else:
        return d
hr_data['New Dept'] = hr_data.department.map(f)
```

```python
hr_data.sample(10)
```

* Giving hike to satisfied people with low salary still with the org

```python
def f(r):
    if (r.satisfaction_level > 0.82) & (r.left == 0) & (r.salary == 'low'):
        return 'medium'
    else:
        return r.salary
    
hr_data['New Salary'] = hr_data.apply(f, axis=1)
```
